```python
# ==================================================
# Additive & Multiplicative Time Series Decomposition
# ==================================================

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# --------------------------------------------------
# 1. Load Dataset
# --------------------------------------------------
df = pd.read_csv("ML471_S1_Datafile_Concept.csv")   # change filename if needed

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)

# Rename columns safely
df.columns = ['Consumption', 'Festival']

series = df['Consumption']

# --------------------------------------------------
# 2. ADDITIVE DECOMPOSITION
# --------------------------------------------------
additive = seasonal_decompose(
    series,
    model='additive',
    period=12   # monthly seasonality
)

plt.figure(figsize=(12,8))

plt.subplot(4,1,1)
plt.plot(additive.observed)
plt.title("Additive Decomposition: Observed")

plt.subplot(4,1,2)
plt.plot(additive.trend)
plt.title("Additive Decomposition: Trend")

plt.subplot(4,1,3)
plt.plot(additive.seasonal)
plt.title("Additive Decomposition: Seasonal")

plt.subplot(4,1,4)
plt.plot(additive.resid)
plt.title("Additive Decomposition: Residual")

plt.tight_layout()
plt.show()

# --------------------------------------------------
# 3. MULTIPLICATIVE DECOMPOSITION
# --------------------------------------------------
multiplicative = seasonal_decompose(
    series,
    model='multiplicative',
    period=12
)

plt.figure(figsize=(12,8))

plt.subplot(4,1,1)
plt.plot(multiplicative.observed)
plt.title("Multiplicative Decomposition: Observed")

plt.subplot(4,1,2)
plt.plot(multiplicative.trend)
plt.title("Multiplicative Decomposition: Trend")

plt.subplot(4,1,3)
plt.plot(multiplicative.seasonal)
plt.title("Multiplicative Decomposition: Seasonal")

plt.subplot(4,1,4)
plt.plot(multiplicative.resid)
plt.title("Multiplicative Decomposition: Residual")

plt.tight_layout()
plt.show()
```
