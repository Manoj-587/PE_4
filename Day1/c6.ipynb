```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load dataset
df = pd.read_csv("ML471_S1_Datafile_Concept(1).csv")

# Convert Datetime column
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)

# Use already differenced column
diff_series = df['Power_Consumption_diff'].dropna()

# Plot ACF & PACF
plt.figure(figsize=(14,5))

plt.subplot(1, 2, 1)
plot_acf(diff_series, lags=30, ax=plt.gca())
plt.title("ACF")

plt.subplot(1, 2, 2)
plot_pacf(diff_series, lags=30, ax=plt.gca(), method='ywm' , color='red')
plt.title("PACF")

plt.tight_layout()
plt.show()
```
