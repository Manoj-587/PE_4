```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ML471_S4_Datafile_Practice.csv", parse_dates=["Date"], index_col="Date")

seasonal_period = 12
seasonal_naive = data["Close"].shift(seasonal_period)

plot_data = pd.DataFrame({
    "Actual": data["Close"],
    "Seasonal_Naive": seasonal_naive
}).dropna()

split = int(len(plot_data) * 0.8)
test = plot_data.iloc[split:]

plt.figure(figsize=(10,5))
plt.plot(test.index, test["Actual"], label="Actual", color="blue")
plt.plot(test.index, test["Seasonal_Naive"], label="Seasonal Naive", color="orange")
plt.xlabel("Date")
plt.ylabel("Closing Stocks")
plt.legend()
plt.tight_layout()
plt.show()
```
